```python
import pandas as pd

# Load the dataset
file_path = dataset_file_path  # make sure 'dataset_file_path' contains the correct file path
data = pd.read_excel(file_path)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs

```

```python
# Generating a synthetic dataset with 3 clusters
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)
```

```python

# Creating a Gaussian Mixture Model with 3 components
gmm = GaussianMixture(n_components=3, random_state=0)
```

```python
# Fitting the model
gmm.fit(X)

# Predicting the cluster labels
labels = gmm.predict(X)
```

```python

# Plotting the clusters and their centroids
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis')
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], s=100, c='red', label='Centroids')
plt.title('Gaussian Mixture Model Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
